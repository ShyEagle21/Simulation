```python


def feature_generator(arrivals_bool, truck_vol_bool, macro_packages_bool, part_packages_bool, LH_packages_bool, USPS_packages_bool, USPN_packages_bool, FDEG_packages_bool, FDE_packages_bool, TLMD_packages_bool):
    features = []
    features.clear()
    features = []
    if arrivals_bool:
        features = features + arrivals
    if truck_vol_bool:
        features = features + truck_vol
    if macro_packages_bool:
        features = features + macro_packages
    if part_packages_bool:
        features = features + part_packages
    if LH_packages_bool:
        features = features + LH_packages
    if USPS_packages_bool:
        features = features + USPS_packages
    if USPN_packages_bool:
        features = features + USPN_packages
    if FDEG_packages_bool:
        features = features + FDEG_packages
    if FDE_packages_bool:
        features = features + FDE_packages
    if TLMD_packages_bool:
        features = features + TLMD_packages
    return features


```
